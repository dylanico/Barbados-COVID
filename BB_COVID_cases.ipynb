```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
```

```python
df = pd.read_excel("owid-covid-data.xlsx")
```

```python
Barbados_data = df[df["location"] == "Barbados"]
```

```python
Barbados_data
```

```python
bb_cases = sns.scatterplot(x=Barbados_data['date'], y = Barbados_data['total_cases'])
spacing = 250
visible = bb_cases.xaxis.get_ticklabels()[::spacing]
for label in bb_cases.xaxis.get_ticklabels():
    if label not in visible:
        label.set_visible(False)


plt.title("Graph showing Barbados' total COVID-19 cases since the beginning of the pandemic")
plt.show()
```
